```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans #This is where we import the package that allows is to do Kmeans clustering
from sklearn import preprocessing as sk_preprocessing
```

```python
#import the data from the webpage into a dataframe called DF
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00236/seeds_dataset.txt", sep= '\s+', header=None)
#Rename columns since they don't come with headers
df.columns=['Area','Perimeter','Compactness','Length of Kernal','Width of Kernel','Asymmetry Coefficient','Length of kernel groove','Class']
```

```python
df.info() #Check out the data to see what's going on
```

```python
data_cluster = df.drop('Class', axis = 1) #Drop the class attribute so the KMeans clustering function can't "Cheat"
```

```python
scaler = sk_preprocessing.StandardScaler() #Turn the data into scalers to get rid of unit discrepincies
data_cluster = scaler.fit_transform(data_cluster)
```

```python
data_cluster #Our new dataframe without the target and appropriately scaled
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 0) #We have 3 classes of seed so we want 3 clusters
```

```python
kmeans.fit(data_cluster) #fit the clusters
```

```python
kmeans.labels_ #look at the clusters and compare to the actual target
```

```python
df['Class'].values #Print the target as an array for comparison
```

```python
#The error is the incorrectly classified records divided by the total
#In this case there are (8+5+4)/210
```

```python
print((8+5+4)/210) #Therefore our error is 8%
```

```python
from sklearn.decomposition import PCA #PCA lets us visualize data in 2-D
pca = PCA(n_components=2)
pca.fit(data_cluster)
```

```python
kmeans_pca = pca.transform(data_cluster) #Transform the data_cluster df with PCA
```

```python
kmeans_pca #PCA stands for principal component analysis.  In this case PCA is used for dimensionality reduction
#which will help us in our following visualizations
```

```python
#Plot the points from the KMeans clustering model.  This plot is showing the different groups that were fitted by the
#Kmeans algorithm
#Look at next picture for the plot of the actual classes
plt.figure(figsize = (10,10))
plt.scatter(kmeans_pca[:,0], kmeans_pca[:,1], c = kmeans.labels_, cmap = 'rainbow')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
#This is the visual for the actual class of seed.  We can compare to the KMeans cluster figure and see they are
#very similar.  However in this case the real groups have some overlap that was not perfectly detected by the Kmeans
#algorithm
plt.figure(figsize = (10,10))
plt.scatter(kmeans_pca[:,0], kmeans_pca[:,1], c = df['Class'], cmap = 'rainbow')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```
